```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mydata = pd.read_csv('NY property data.csv')
```

```python
mydata.head()
```

```python
mydata['EASEMENT'].value_counts()
```

```python
plt.rcParams["figure.figsize"] = [15,8]
sns.set(font_scale=2)
plot = sns.countplot(x='EASEMENT', data = mydata, order = mydata['EASEMENT'].value_counts().index)
plot.set_yscale('log')
plt.title('Bar Plot of "EASEMENT" Field')
plt.ylabel('log Scale of Count')
for p in plot.patches:
    height = p.get_height()
    plot.text(p.get_x()+p.get_width()/2., height + 0.1,
        height,ha="center",alpha=0.6,size=16)
plt.savefig('hist.png')
```
